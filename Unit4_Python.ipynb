iii. Operators
    -Arithmetic
    -Logical
    -Comparison
    -Bitwise

```python
# A) Arithmetic Operators

# Operator	          Description	             Example

# +	                  Addition	                 5 + 3 â†’ 8
# -	                  Subtraction	             5 - 3 â†’ 2
# *	                  Multiplication	         5 * 3 â†’ 15
# /                   Division	                 5 / 2 â†’ 2.5
# //	              Floor Division	         5 // 2 â†’ 2
# %	                  Modulus (Remainder)	     5 % 2 â†’ 1
# **	              Exponentiation	         2 ** 3 â†’ 8


#Example code:
a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.333
print(a // b) # 3
print(a % b)  # 1
print(a ** b) # 1000
```

```python
# B) Logical Operators

# Operator       	Description                      	              Example

# and	            Returns True if both statements are true       	  (5 > 3) and (3 > 1) â†’ True
# or	            Returns True if one of the statements is true	  (5 > 3) or (3 < 1) â†’ True
# not	            Reverses the result	not                           (5 > 3) â†’ False


#Example code
x = 5
y = 3

print((x > 2) and (y > 1))  # True
print((x > 2) or (y < 1))   # True
print(not (x > y))         # False
```

```python
# C) Comparison Operators

# Operator	   Description	                   Example

# ==	       Equal to	                       5 == 3 â†’ False
# !=	       Not equal to	                   5 != 3 â†’ True
# >	           Greater than	                   5 > 3 â†’ True
# <	           Less than	                   5 < 3 â†’ False 
# >=	       Greater than or equal to	       5 >= 5 â†’ True
# <=	       Less than or equal to	       5 <= 3 â†’ False 


#Example code
x = 5
y = 3

print(x == y)  # False
print(x != y)  # True
print(x > y)   # True
print(x < y)   # False
print(x >= y)  # True
print(x <= y)  # False
```

```python
# D) Bitwise Operators 

# Operator	       Description	           Example

# &	               Bitwise AND	           5 & 3 â†’ 1
# |	               Bitwise OR              5 | 3 â†’ 7
# ^	               Bitwise XOR	           5 ^ 3 â†’ 6
# ~	               Bitwise NOT	           ~5 â†’ -6
# <<	           Left Shift	           5 << 1 â†’ 10
# >>	           Right Shift	           5 >> 1 â†’ 2


#Example code
a = 5  # 101 in binary
b = 3  # 011 in binary

print(a & b)  # 1 (001)
print(a | b)  # 7 (111)
print(a ^ b)  # 6 (110)
print(~a)     # -6 (bitwise NOT)
print(a << 1) # 10 (shifts left by 1 bit)
print(a >> 1) # 2 (shifts right by 1 bit)
```

i. Conditional Statements (if, elif, else)

```python
#Syntax 

if condition:
    # Code runs if condition is True
elif another_condition:
    # Code runs if previous condition is False but this one is True
else:
    # Code runs if all conditions are False
```

3. DATA STRUCTURES - (list , tuple, set , dictionary)

i. List

```python
#list - stores elements of multiple data types
#array - stores elements of same data types

#Creating a List
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = [10, "hello", 3.14]

#Accessing Elements
print(fruits[0])      # apple (first element)
print(fruits[-1])     # cherry (last element)

#Modifying a List
fruits[1] = "mango"  
print(fruits)         # ['apple', 'mango', 'cherry']

#Adding Elements
fruits.append("orange")    # Adds at the end
fruits.insert(1, "grape")  # Inserts at index 1
print(fruits)

#Removing Elements
fruits.remove("mango")  # Removes by value
fruits.pop(2)           # Removes by index (default: last element)
del fruits[0]           # Deletes first element
print(fruits)

#Looping Through a List
for fruit in fruits:
    print(fruit)
```

ii. Tuple

```python
#Creating a Tuple
#Once a tuple is created, you cannot change, add, or remove its elements

coordinates = (10, 20)
fruits = ("apple", "banana", "cherry")

#Accessing Elements
print(fruits[1])  # banana

#Tuple Unpacking
x, y = coordinates  
print(x)  # 10
print(y)  # 20
```

iii. set

```python
#A set is an unordered collection of unique elements.

#Creating a Set
my_set = {1, 2, 3, 4, 4, 5}  # Duplicate values are removed
print(my_set)                # {1, 2, 3, 4, 5}

#Adding & Removing Elements
my_set.add(6)  
my_set.remove(3)
print(my_set)

#Set Operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1 | set2)  # Union: {1, 2, 3, 4, 5}
print(set1 & set2)  # Intersection: {3}
print(set1 - set2)  # Difference: {1, 2}
```

iv. dict

```python
#A dictionary stores key-value pairs.

#Creating a Dictionary
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

#Accessing Values
print(student["name"])  # Alice
print(student.get("age"))  # 20

#Adding & Modifying Values
student["age"] = 21  
student["city"] = "New York"
print(student)

#Removing Elements
del student["grade"]
print(student)

#Looping Through a Dictionary
for key, value in student.items():
    print(key, ":", value)
```

v. List & Dictionary Comprehensions

```python
#Python provides a concise way to create lists and dictionaries.

#List Comprehension
numbers = [x for x in range(1, 6)]  # [1, 2, 3, 4, 5]
squares = [x**2 for x in range(1, 6)]  # [1, 4, 9, 16, 25]
even_numbers = [x for x in range(10) if x % 2 == 0]  # [0, 2, 4, 6, 8]


#Dictionary Comprehension
squares_dict = {x: x**2 for x in range(1, 6)}
print(squares_dict)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

4. FUNCTIONS AND MODULES

i. Defining Functions (def keyword)

```python
#A function is a block of code that runs when called.

#Basic Function
def greet():
    print("Hello, Siddhi!")  
greet()                   # Calling the function


#Function with Parameters
def greet(name):
    print(f"Hello, {name}!")
greet("Siddhi")           # Output: Hello, Siddhi!


#Function with Return Value
def add(a, b):
    return a + b
result = add(5, 3)
print(result)             # Output: 8
```

ii. Arguments (*args and **kwargs)

```python
#Python allows flexible function arguments.


#Positional Arguments (*args)
#Used to pass multiple values as a tuple.

def add_numbers(*args):
    return sum(args)

print(add_numbers(1, 2, 3, 4))  # Output: 10


#Keyword Arguments (**kwargs)
#Used to pass multiple key-value pairs as a dictionary.

def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

show_info(name="Siddhi", age=20, city="Ahmedabad")

# Output:
# name: Siddhi
# age: 20
# city: Ahmedabad
```

iii. Lambda Functions

```python
#Syntax 
# lambda arguments: expression

#Examples:
square = lambda x: x**2
print(square(5))  # Output: 25

add = lambda a, b: a + b
print(add(3, 7))  # Output: 10

#Using lambda in Sorting
students = [("Alice", 20), ("Bob", 18), ("Charlie", 22)] #list of tuples
students.sort(key=lambda x: x[1])  # Sort by age
print(students)
```

ðŸš€ Mini Project - 1: "Smart Calculator" (Using Functions & Modules)

ðŸ“Œ Project Overview
We will create a Python program that:
âœ… Takes user input for a mathematical operation (addition, subtraction, multiplication, etc.).
âœ… Uses functions for calculations.
âœ… Accepts multiple numbers using *args.
âœ… Uses the math module for advanced operations like square root and power.
âœ… Provides an interactive menu.

```python
import math  # Importing math module

# Function for addition
def add(*args):
    return sum(args)

# Function for subtraction
def subtract(a, b):
    return a - b

# Function for multiplication
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

# Function for division
def divide(a, b):
    return a / b if b != 0 else "Cannot divide by zero"

# Function for square root
def square_root(num):
    return math.sqrt(num)

# Function for power
def power(base, exponent):
    return math.pow(base, exponent)

# Function for calculator menu
def calculator():
    print("\nðŸ“Œ Welcome to Smart Calculator!")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Square Root")
    print("6. Power")
    print("7. Exit")
    
    while True:
        choice = input("\nEnter your choice (1-7): ")
        
        if choice == "1":
            nums = list(map(float, input("Enter numbers separated by space: ").split()))
            print("Result:", add(*nums))
        
        elif choice == "2":
            a, b = map(float, input("Enter two numbers: ").split())
            print("Result:", subtract(a, b))
        
        elif choice == "3":
            nums = list(map(float, input("Enter numbers separated by space: ").split()))
            print("Result:", multiply(*nums))
        
        elif choice == "4":
            a, b = map(float, input("Enter two numbers: ").split())
            print("Result:", divide(a, b))
        
        elif choice == "5":
            num = float(input("Enter a number: "))
            print("Square Root:", square_root(num))
        
        elif choice == "6":
            base, exponent = map(float, input("Enter base and exponent: ").split())
            print("Result:", power(base, exponent))
        
        elif choice == "7":
            print("Exiting the calculator. Goodbye! ðŸ‘‹")
            break
        
        else:
            print("Invalid choice! Please enter a number between 1-7.")

# Run the calculator
calculator()
```

ðŸš€ Mini Project- 2: To-Do List App (Using Functions & Lists)

ðŸ“Œ Project Overview
We will create a simple text-based To-Do List App that allows users to:
âœ… Add tasks to the list.
âœ… View tasks in an organized manner.
âœ… Mark tasks as done (remove from the list).
âœ… Exit the app when finished.

```python
# To-Do List App

# Initialize an empty task list
tasks = []

# Function to add a task
def add_task(task):
    tasks.append(task)
    print(f"âœ… Task '{task}' added successfully!")

# Function to view all tasks
def view_tasks():
    if not tasks:
        print("ðŸ“Œ No tasks in the list.")
    else:
        print("\nðŸ“ Your To-Do List:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")

# Function to remove a completed task
def remove_task(index):
    if 0 < index <= len(tasks):
        removed_task = tasks.pop(index - 1)
        print(f"âœ… Task '{removed_task}' marked as done and removed!")
    else:
        print("âŒ Invalid task number!")

# Function to run the To-Do List App
def todo_list():
    print("\nðŸ“Œ Welcome to Your To-Do List!")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Completed Task")
    print("4. Exit")

    while True:
        choice = input("\nEnter your choice (1-4): ")

        if choice == "1":
            task = input("Enter task: ")
            add_task(task)

        elif choice == "2":
            view_tasks()

        elif choice == "3":
            view_tasks()
            try:
                index = int(input("Enter the task number to mark as done: "))
                remove_task(index)
            except ValueError:
                print("âŒ Please enter a valid number!")

        elif choice == "4":
            print("ðŸš€ Exiting To-Do List. Stay productive! ðŸ‘‹")
            break

        else:
            print("âŒ Invalid choice! Please enter a number between 1-4.")

# Run the To-Do List App
todo_list()
```
